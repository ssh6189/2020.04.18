```python
#현재
import cv2
import time
import numpy as np
import datetime
from tkinter import *
from PyQt5 import QtCore
from PyQt5 import QtWidgets
from PyQt5 import QtGui
import sys
from os import path
from PyQt5.QtWidgets import QApplication, QWidget, QLabel, QPushButton
        
c = 0

def nt():
    global c
    if(c == 5):
        c = c
    else:
        c = c + 1

        
def bk():
    global c
    if(c == 0):
        c = 0
    else:
        c = c - 1


def remote():
    rm = Tk()
    rm.title("Webcam_Controller")
    rm.geometry("300x300")

    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    record = False

    nexts = Button(rm, text="next", command=nt)
    nexts.grid(row = 1, column = 2)
    backs = Button(rm, text="back", command=bk)
    backs.grid(row = 1, column = 1)

    rm.mainloop()

fourcc = cv2.VideoWriter_fourcc(*'XVID')
record = False

logo = cv2.imread("./zero.png")

#logo = cv2.resize(logo, (275,183))

#logo_height, logo_width, _ = logo.shape

cap = cv2.VideoCapture(0)

cap.set(10000, 10000)

#cap_height = cv2.CAP_PROP_FRAME_HEIGHT
#cap_width =  cv2.CAP_PROP_FRAME_WIDTH

time.sleep(3) #warming up
if not cap.isOpened():
    exit()

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

remote()

while True:
    
    ret, image = cap.read()
    
    now = datetime.datetime.now().strftime("%d_%H-%M-%S")
    key = cv2.waitKey(3)
    
    if key == 51: #3
        if(c == 2):
            c = c
        else:
            c = c + 1
    elif key == 49: #1
        if(c == 0):
            c = 0
        else:
            c = c - 1
    elif key == 48: #0
        c = 0
        
    if c == 0: 
        logo = cv2.imread("./zero.png")
    elif c == 1: 
        logo = cv2.imread("./pick.png")
    elif c == 2: 
        logo = cv2.imread("./smile.png")
    elif c == 3:
        logo = cv2.imread("./keobi.png")
    elif c == 4:
        logo = cv2.imread("./hahoi.png")
    elif c == 5:
        logo = cv2.imread("./trumph.png")


    faces = face_cascade.detectMultiScale(image)
    for face in faces:
        (x, y, w, h) = face
        x1, y1, x2, y2 = x, y, x + w, y + h
        #cv2.rectangle(image, (x1, y1), (x2, y2), (0, 0, 255), thickness=2)

        image_face = image[y1:y2, x1:x2]
        #logo = cv2.resize(logo, (y1-y2, x1-x2))
        #image_face = logo[y1:y2, x1:x2]
        logo = cv2.resize(logo, (w, h))
        image_face =  cv2.add(image_face, logo)
        image_face = cv2.resize(image_face, (w, h), interpolation=cv2.INTER_AREA)
        image[y1:y2, x1:x2] = image_face
        
        if key == 122: #z
            #print("캡쳐")
            cv2.imwrite("./" + str(now) + ".png", image)
        elif key == 120: #x
            #print("녹화 시작")
            record = True
            video = cv2.VideoWriter("./" + str(now) + ".avi", fourcc, 20.0, (image.shape[1], image.shape[0]))
        elif key == 121: #c
            #print("녹화 중지")
            record = False
            video.release()

        if record == True:
            #print("녹화 중..")
            video.write(image)
            
        
        #remote()
        cv2.imshow("result", image)
        
    if key == 113: #q
        break

    if not ret:
        break

    #image = cv2.add(image, logo)

cap.release()
cv2.destroyAllWindows()
```
